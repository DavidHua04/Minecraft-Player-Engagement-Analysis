```python
# things to do before
import pandas as pd
import altair as alt

alt.data_transformers.enable('vegafusion')
```

```python
players=pd.read_csv('players.csv')
players
```

```python
players_filtered=players.drop(columns=['hashedEmail', 'individualId', 'organizationName', 'name'])
players_filtered
```

```python
played_hours_vs_age_plot=alt.Chart(players_filtered).mark_point().encode(
    x=alt.X('age')
        .title('Age'),
    y=alt.Y('played_hours')
        .title('Played Hours'),
    color=alt.Color('age')
        .title('Age')
        .scale(scheme='category20c')
).properties(width=700)
played_hours_vs_age_plot
```

```python
age_count_plot=alt.Chart(players_filtered).mark_bar().encode(
    x=alt.X('age').title('Age'),
    y=alt.Y('count()')
)
age_count_plot
```
